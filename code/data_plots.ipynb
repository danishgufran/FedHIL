```python
import re
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt

def calculate_zipf_prob(word_rank, s):
    zipf_probs = []

    # Calculate the frequencies using the formula f(n) = C / n^s
    frequencies = [(word, rank, 1 / (rank ** s)) for word, _, rank in word_rank]

    # Calculate the normalization constant by summing the frequencies
    normalization_constant = sum([freq for _, _, freq in frequencies])

    # Normalize the frequencies to get probabilities
    zipf_probs = [(word, rank, freq / normalization_constant) for word, rank, freq in frequencies]

    return zipf_probs

# Read the text from the file
filename = 'resources.txt'
with open(filename, 'r') as file:
    text = file.read()

# Preprocess the text
text = text.lower()
words = re.findall(r'\b\w+\b', text)

# Count the frequency of each unique word
word_freq = Counter(words)

# Sort the unique words in descending order of frequency
sorted_word_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

# Assign a rank to each unique word
word_rank = [(word, freq, rank + 1) for rank, (word, freq) in enumerate(sorted_word_freq)]

# Calculate the Zipf's distribution for each value of s
s_values = [0.6, 0.7, 0.8, 0.9]
zipf_distributions = {s: calculate_zipf_prob(word_rank, s) for s in s_values}

# Print the results
for s, zipf_distribution in zipf_distributions.items():
    print(f"s: {s}")
    for word, rank, prob in zipf_distribution:
        print(f"Word: {word}, Rank: {rank}, Probability: {prob}")
    print("\n")




plt.figure(figsize=(12, 4))
for s, zipf_distribution in zipf_distributions.items():
    ranks = [rank for _, rank, _ in zipf_distribution]
    probabilities = [prob for _, _, prob in zipf_distribution]
    plt.plot(ranks, probabilities, label=f's: {s}')


plt.xlabel('Rank')
plt.ylabel('Probability')
plt.title("Zipf's Distribution for Different Values of s")
plt.xticks(range(0, 176, 25))
# plt.yticks(np.arange(0, 1, 0.1))

plt.legend()
# Set the background color to black and the grid color to white
plt.rcParams['axes.facecolor'] = 'white'
plt.grid(color='black', linestyle='-', linewidth=0.2)
# plt.savefig('zipf_distribution.png', dpi=300)

# Display the plot
plt.show()
```
